```python
categorical_encoded = pd.get_dummies(categorical)
categorical_encoded
```

```python
data_final = pd.concat([numerical, categorical_encoded], axis=1)
data_final
```

```python
data_final.columns
```

# Splitting the in Training and testing Sets

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import log_loss
from sklearn.metrics import auc
from sklearn.metrics import f1_score
```

```python
target = data['Attrition']
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_final, target, train_size = 0.8, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Model Creation

## Logistic Regression

```python
lr_model  = LogisticRegression()
lr_model.fit(x_train, y_train)
lr_model_predictions = lr_model.predict(x_test)

print('Accuracy          : ', accuracy_score(y_test, lr_model_predictions))
print('Log-Loss          : ', log_loss(y_test, lr_model_predictions))
print('F1-score          : ', metrics.f1_score(y_test, lr_model_predictions, average='weighted'))
#print('Arear-Under-Curve : ', metrics.auc(y_test, lr_model_predictions))
print("Classification Report :\n",classification_report(y_test, lr_model_predictions))
```

## Decision Tree Classifier

```python
dtc_model  = DecisionTreeClassifier()
dtc_model.fit(x_train, y_train)
dtc_model_predictions = dtc_model.predict(x_test)

print('Accuracy          : ', accuracy_score(y_test, dtc_model_predictions))
print('Log-Loss          : ', log_loss(y_test, dtc_model_predictions))
print('F1-score          : ', metrics.f1_score(y_test, dtc_model_predictions, average='weighted'))
#print('Arear-Under-Curve : ', metrics.auc(y_test, lr_model_predictions))
print("Classification Report :\n",classification_report(y_test, dtc_model_predictions))
```

# Random Forest Classifier

```python
rfc_model  = RandomForestClassifier()
rfc_model.fit(x_train, y_train)
rfc_model_predictions = rfc_model.predict(x_test)

print('Accuracy          : ', accuracy_score(y_test, rfc_model_predictions))
print('Log-Loss          : ', log_loss(y_test, rfc_model_predictions))
print('F1-score          : ', metrics.f1_score(y_test, rfc_model_predictions, average='weighted'))
#print('Arear-Under-Curve : ', metrics.auc(y_test, lr_model_predictions))
print("Classification Report :\n",classification_report(y_test, rfc_model_predictions))
```
